# NLP_04: Removing Stopwords with NLTK in Python

In natural language processing (NLP), it is often useful to remove commonly occurring words that carry little semantic meaning. These words, called **stopwords**, include articles, prepositions, and pronouns such as *"the"*, *"and"*, *"is"*, and *"in"*. While they may seem insignificant, handling stopwords properly can significantly improve the accuracy and efficiency of NLP tasks.

### Example

| Original Text                                       | Without Stopwords                              |
| --------------------------------------------------- | ---------------------------------------------- |
| A Computer Science Portal for Geeks | Computer Science, Portal, Geeks |
| Can listening be exhausting?                        | Listening, Exhausting                          |
| I like reading, so I read                           | Like, Reading, read                            |

Consider the sentence:
*"The quick brown fox jumps over the lazy dog"*

* **Stopwords:** "the", "over"
* **Content words:** "quick", "brown", "fox", "jumps", "lazy", "dog"

Removing stopwords becomes particularly important when working with large text datasets, as processing high-frequency words can waste computational resources and skew analysis results.

### When to Remove Stopwords

The decision depends on the NLP task:

**Tasks that benefit from stopword removal:**

* Text classification and sentiment analysis
* Information retrieval and search engines
* Topic modeling and clustering
* Keyword extraction

**Tasks that require keeping stopwords:**

* Machine translation (maintains grammatical structure)
* Text summarization (preserves sentence coherence)
* Question-answering systems (syntactic relationships matter)
* Grammar checking and parsing

Language modeling is a middle ground: whether to remove stopwords depends on the specific application and computational resources.


### Categories of Stopwords

Understanding different types of stopwords helps make informed decisions:

1. **Standard Stopwords:** Common function words like articles (*"a"*, *"the"*), conjunctions (*"and"*, *"but"*), and prepositions (*"in"*, *"on"*).
2. **Domain-Specific Stopwords:** Frequently occurring words specific to a field, e.g., *"patient"* in medical texts.
3. **Contextual Stopwords:** Words that appear very often in a particular dataset.
4. **Numerical Stopwords:** Digits, punctuation, or single characters.


### Removing Stopwords Using NLTK

NLTK (Natural Language Toolkit) supports stopword removal in **16 languages**. The typical process includes:

1. **Setup:** Import NLTK modules and download resources like stopwords and tokenizers.
2. **Text Preprocessing:** Convert text to lowercase and tokenize it into words.
3. **Stopword Removal:** Load English stopwords and filter them out from the token list.
4. **Output:** Compare original and cleaned tokens.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# Download necessary NLTK resources (only needed once)
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# Sample text
txt = """
    I recently bought this phone from Amazon, and honestly, it’s not that great.
    The battery life is poor, the camera is okay, but the price is way too high.
    I wouldn’t recommend it to anyone!
    """
```

```python
# Tokenize text
words = word_tokenize(txt.lower())
stop_words = set(stopwords.words("english"))
```

```python
# Remove stopwords
filtered_txt = [word for word in words if word not in stop_words]
```

```python
print(txt)
```

```python
# Print the filtered text
print(filtered_txt)
```
